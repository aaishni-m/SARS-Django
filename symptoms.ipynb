```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("symptoms.csv")
```

```python
data.sum().duplicated
```

```python
def assign_severity(row):
    if row['Severity_Mild'] == 1:
        return "Mild"
    elif row['Severity_Moderate'] == 1:
        return "Moderate"
    elif row['Severity_None'] == 1:
        return "None"
    elif row['Severity_Severe'] == 1:
        return "Severe"
    else:
        return None
```

```python
data['Severity'] = data.apply(assign_severity, axis=1)
```

```python
data
```

```python
def assign_gender(row): 
    if row['Gender_Transgender'] == 1:
        return 'Other'
    elif row['Gender_Male'] == 1:
        return 'Male'
    elif row['Gender_Female'] == 1:
        return 'Female'
    else:
        return 'NA'
```

```python
data['Gender'] = data.apply(assign_gender, axis = 1)
```

```python
def assign_age(row):
    if row['Age_0-9'] == 1:
        return '0-9'
    elif row['Age_10-19'] == 1:
        return '10-19'
    elif row['Age_20-24']==1:
        return '20-24'
    elif row['Age_25-59']==1:
        return '25-59'
    elif row['Age_60+']==1:
        return '60+'
    else:
        return "-"
```

```python
data['Age'] = data.apply(assign_age, axis=1)
```

```python
data.drop(columns=['Severity_Mild', 'Severity_Moderate','Severity_None', 'Severity_Severe','Gender_Transgender', 'Gender_Male', 'Gender_Female','Contact_Dont-Know','Contact_No','Contact_Yes','Country'], inplace=True)
```

```python
data = data.drop(columns=['Age_60+','Age_25-59','Age_20-24','Age_10-19','Age_0-9'])
```

```python
# Add oxygen levels based on severity
def generate_oxygen_level(row):
    if row['Severity'] == "None":
        return np.random.uniform(98, 100) 
    elif row['Severity'] == "Mild":
        return np.random.uniform(94, 100) 
    elif row['Severity'] == "Moderate":
        return np.random.uniform(88, 94)  
    elif row['Severity'] == "Severe":
        return np.random.uniform(75, 88)  
    else:
        return 92

# Apply the function to each row
data['Oxygen Level'] = data.apply(generate_oxygen_level, axis=1)

```

```python
data
```

```python
def classify_covid(row):
    if row['Severity'] in ['Mild', 'Moderate', 'Severe'] and (
        row['Fever'] == 1 or row['Tiredness'] == 1 or row['Dry-Cough'] == 1 or row['Difficulty-in-Breathing'] == 1):
        return 'COVID'
    else:
        return 'Non-COVID'
```

```python
data['Status'] = data.apply(classify_covid, axis=1)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
x = data.drop(columns=['Status'])
y = data['Status']
```

```python
xtest, xtrain, ytest, ytrain = train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
model = LogisticRegression()
```

```python
model.fit(xtrain,ytrain)
```

```python
ypred = model.predict_proba(xtest)
```

```python
probabilities = model.predict_proba(xtest)[:, 1]  # Probability of the positive class (COVID)

# Format probabilities as percentages
probabilities_percentage = [f"{prob * 100:.2f}%" for prob in probabilities]
```

```python
for i, prob in enumerate(probabilities_percentage[:10]):
    print(f"Patient {i+1}: {prob} chance of having COVID")
```
